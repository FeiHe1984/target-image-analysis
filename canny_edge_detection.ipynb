# Canny Edge Detection

In this notebook, I'm going to explore feasibility of canny edge detection for selecting different features in a target.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

# Set the default figure size
plt.rcParams['figure.figsize'] = [16,8]

# read in the image
image = cv2.imread('images/template.jpeg')

# change color to RGB from BGR
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# show image
plt.imshow(image)
```

```python
# convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# show image
plt.imshow(gray, cmap='gray')
```

### Implement Canny Edge Detectcion

```python
# "wide" and "tight" thresholds
wide = cv2.Canny(gray, 20, 100)
tight = cv2.Canny(gray, 200, 240)

# show images
figure, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,10))

ax1.set_title('wide threshold')
ax1.imshow(wide, cmap='gray')

ax2.set_title('tight threshold')
ax2.imshow(tight, cmap='gray')
```

It seems canny edge detection will not be usefull for selecting features on the template.

```python
# "wide" and "tight" thresholds
#wide = cv2.Canny(gray, 20, 100)
#tight = cv2.Canny(gray, 200, 240)

# Change the "wide" and "tight" thresholds
wide = cv2.Canny(gray, 0, 100)
tight = cv2.Canny(gray, 220, 250)

# show images
figure, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,10))

ax1.set_title('wide threshold')
ax1.imshow(wide, cmap='gray')

ax2.set_title('tight threshold')
ax2.imshow(tight, cmap='gray')
```

The algorithm will first detect strong pixels above the high_threshold, and reject pixels below the low_threshold. Next, pixels with values between the low_threshold and high_threshold will be included as long as they are connected to strong edges. The output edges is a binary image with white pixels tracing out the detected edges and black everywhere else.
