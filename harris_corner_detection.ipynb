# Harris Corner Detection

In this notebook, I'm going to detect corners of a taget.

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline

# set the default figure size
plt.rcParams['figure.figsize'] = [16,8]

# read in the image
image = cv2.imread('images/template.jpeg')

# make copy of image
image_copy = np.copy(image)

# change color to RGB from BGR
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

# show image
plt.imshow(image_copy)
```

### Detect Corners

```python
# convert to grayscale
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)
gray = np.float32(gray)

# detect corners
corners = cv2.cornerHarris(gray, 30, 5, 0.05)

# dilate corner image to enhance corner points
corners = cv2.dilate(corners, None)

# show image
plt.imshow(corners, cmap='gray')
```

### Extract and Display Strong Corners

```python
# threshold for extracting strong corners
thresh = 0.1*corners.max()

# create image copy to draw corners on
corner_image = np.copy(image_copy)

# iterate through all corners and draw them
for j in range(0, corners.shape[0]):
    for i in range(0, corner_image.shape[1]):
        if (corners[j,i] > thresh):
            # image, center point, radius, colorm thickness
            cv2.circle(corner_image, (i,j), 1, (0,255,0), 1)
            
# show image
plt.imshow(corner_image)
```

The corners of the target got detected very nicely.
